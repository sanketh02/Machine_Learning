```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import accuracy_score, classification_report, roc_curve, roc_auc_score
import joblib
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading data set

path = "C:\\Users\\123\Desktop\\python project for data analitycs\\Data\\drug200.csv"

data = pd.read_csv(path)
```

```python
data.head()
```

```python
df = data.copy()
```

```python
# Encoding
le = LabelEncoder()
```

```python
df["BP"] = le.fit_transform(df["BP"])
df["Cholesterol"] = le.fit_transform(df["Cholesterol"])
df["Drug"] = le.fit_transform(df["Drug"])
df["Sex"] = le.fit_transform(df["Sex"])
```

```python
x = df.drop("Drug",axis=1)
y = df["Drug"]
```

```python
x.head()
```

```python
y.head()
```

```python
print("Shape of Feature data and Target data :", x.shape,y.shape)
```

```python
# Spliting data into train test case

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2 , random_state = 40)
```

```python
x_train.shape , x_test.shape , y_train.shape, y_test.shape
```

```python
scl = StandardScaler()
```

```python
x_train_scl = scl.fit_transform(x_train)
x_test_scl = scl.transform(x_test)
```

```python
Log_R = LogisticRegression()
```

```python
Log_R.fit(x_train_scl,y_train)
```

```python
Log_R.intercept_
```

```python
y_pred = Log_R.predict(x_test_scl)
```

```python
pd.DataFrame({"Y-Test":y_test,"Y-Pred":y_pred})[:10]
```

```python
accuracy_score(y_test,y_pred)
```

```python
print("="*80)
print("Classification Metrics")
print()
print(classification_report(y_test,y_pred))
print("="*80)
```

```python
joblib.dump(Log_R,"Logistic_Reg1.pkl")
```
