```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.feature_selection import SelectKBest, f_regression
import joblib
import warnings
warnings.filterwarnings('ignore')
```

```python
print("==========KNN Model============")
```

```python
# Step 1 : Using California housing dataset (bilt in sklearn datasets)
data = fetch_california_housing()
```

```python
x = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
y = pd.Series(data.target , name= "target")
```

```python
print(x.head())
print()
print(y.head())
```

```python
# checking and duplicated values
x.isnull().sum()
```

```python
x.duplicated().sum()
```

```python
x.dtypes
```

```python
x.describe()
```

```python
# Splitting data into test train data

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42, stratify = None)
```

```python
x_train.shape , x_test.shape, y_train.shape, y_test.shape
```

```python
# fitting x_train and x_test into scalar matrix

scaler = StandardScaler()
```

```python
x_train_scl = scaler.fit_transform(x_train)
x_test_scl = scaler.fit_transform(x_test)
```

```python
x_train_scl.shape
```

```python
x_train_scl
```

```python
# Loading Model 
knn = KNeighborsRegressor(n_neighbors=3)
```

```python
# Training model

knn.fit(x_train_scl,y_train)
```

```python
knn.score(x_test_scl,y_test)
```

```python
# checking prediction of model using X_test_scl

y_pred = knn.predict(x_test_scl)
```

```python
y_pred
```

```python
# calculatig Mean squared error and mean absolute error and r2 score

print("Mean Squared error :", mean_squared_error(y_test,y_pred))
print("=" * 50)
print("Mean Absolute error :", mean_absolute_error(y_test,y_pred))
print("="*50)
print("r2 Score :" , r2_score(y_test,y_pred))
```
