```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error,root_mean_squared_error, r2_score
import joblib
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading data
path = "C:\\Users\\123\\Desktop\\python project for data analitycs\\Data\\Advertising.csv"

data = pd.read_csv(path)
```

```python
data.head()
```

```python
df = data.iloc[:,1:]
```

```python
df.describe()
```

```python
x = df.drop("Sales",axis=1)
y = df["Sales"]
```

```python
x.shape, y.shape
```

```python
# Splitting data into test train case

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 45)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
scl = StandardScaler()
```

```python
x_train_scl = scl.fit_transform(x_train)
x_test_scl = scl.transform(x_test)
```

```python
x_train_scl
```

```python
x_test_scl
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(x_train_scl,y_train)
```

```python
Lr.intercept_
```

```python
Lr.coef_
```

```python
y_pred = Lr.predict(x_test_scl)
```

```python
pd.DataFrame({"Y_pred":y_pred,"Y_test":y_test})
```

```python
# Model evaluation or performance matrix
print()
print(f"Mean Squared Error : {mean_squared_error(y_test,y_pred)}")
print()
print("="*60)
print()
print(f"Root Mean Squared Error : {root_mean_squared_error(y_test,y_pred)}")
print()
print("="*60)
print()
print(f"Mean Absolute error : {mean_absolute_error(y_test,y_pred)}")
print()
print("="*60)
print()
print(f"R2 Score : {r2_score(y_test,y_pred)}")
print()
print("="*60)
print()
r2 = r2_score(y_test,y_pred)
n = x_test.shape[0]
k = x_test.shape[1]
adjusted_r2 = 1 - ((1-r2)*(n-1))/(n-k-1)
print(f"Adjusted R2 : {adjusted_r2}") 
```

```python
joblib.dump(Lr, "linear.pkl")
```
