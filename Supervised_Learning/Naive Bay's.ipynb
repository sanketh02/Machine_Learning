```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
```

```python
# Loading Data
data = pd.read_csv("_MConverter.eu_Placement Data.csv")
```

```python
data.head()
```

```python
df = data.iloc[:,1:]
```

```python
df.head()
```

```python
le = LabelEncoder()
```

```python
df["grade"] = le.fit_transform(df["grade"])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# defining feature X and target Y

x = df.drop("placement",axis = 1)
y = df["placement"]
```

```python
x
```

```python
y
```

```python
#splitting data into test train case

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
# Standardization( mean = 0 and sd = 1)

scaler = StandardScaler()
```

```python
x_train_scl = scaler.fit_transform(x_train)
x_test_scl = scaler.fit_transform(x_test)
```

```python
x_train_scl
```

```python
x_test_scl
```

```python
NB_model = GaussianNB()
```

```python
NB_model.fit(x_train_scl,y_train)
```

```python
y_pred = NB_model.predict(x_test_scl)
```

```python
y_pred
```

```python
# Calculating Accuracing score
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score :" , accuracy)
```

```python
# Classification Report
print(" ==========\n Classification Report ========= ")
print(classification_report(y_test,y_pred))
```

```python
NB_model.score(x_test_scl,y_test)
```

```python
pd.DataFrame({"y_true": y_test, "y_pred": y_pred})
```

```python
# Tune hyperparameter
# Define parameter grid for GaussianNB (var_smoothing is the main parameter to tune)

param_grid = {
    'var_smoothing': np.logspace(0, -9, num=10)
}
param_grid
```

```python
# Performance grid search
grid_search = GridSearchCV(GaussianNB(), param_grid, cv=5, scoring='accuracy')

grid_search.fit(x_train_scl, y_train)
```

```python
best_model = grid_search.best_estimator_

print("\n Best Parameter :", grid_search.best_params_)
print("Best Cross_Validation Score: " ,grid_search.best_score_)
```

```python
# Step 9: Evaluate tuned model
# Make predictions with tuned model
y_pred_tunned = best_model.predict(x_test_scl)
```

```python
tunned_accuracy = accuracy_score(y_test,y_pred_tunned)
tunned_accuracy
```

```python
best_model.score(x_test_scl,y_test)
```
